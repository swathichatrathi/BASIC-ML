# Visualization:

# COUNT PLOT

```python
plt.title('No. of Species')
plot=sns.countplot(x="species",data=df)
```

From this visualization,we can observe that our dataset is not balanced as we don't have equal number of count of the species.
We can also observe that we have more number of Adelie species of the penguins maintaining a huge differnce with the count of the Chinstrap.

# SCATTER PLOT

```python
plot=sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',hue="species",data=df)
```

From the above visualization, we can tell that the Adelie species has smaller bill length but higher depth. 
While we see Chinstrap species lies in almost middle for length as well as depth. 
And the Gentoo species has larger bill lengths and smaller bill depths.

# STRIP PLOT

```python
fig, axs = plt.subplots(figsize = (6,6))
plot=sns.stripplot(x="species",y="bill_length_mm",hue="sex",data=df);
```

From the above visualization,we observed that the male penguins has higher bill length when compared to female penguins in all the three species of our dataset.

# BOX PLOT

```python
fig, axs = plt.subplots(figsize = (8,6))
plot=sns.boxplot(x="species",y="body_mass_g",data=df)
```

Here,we can observe that both Adelie and Chinstrap species have almost equal range of body masses.So,we can only differentiate the Gentoo species from the above box plot.We can also observe the outliers present among the Chinstrap species.

## Correlation Between Variables

```python
df.corr()
```

Further, we can visualise this correlation using a heatmap.


# Heat map

```python
fig = plt.figure(figsize = (15,9))
sns.heatmap(df.corr(), cmap='Reds', annot = True);
```

From the above heatmap, we see that flipper_length and body mass have a high correlation,
body mass and bill length have good correlation as well as flipper_length and bill_length have good correlations.

# HIST PLOT

```python
df.hist(edgecolor='black',bins = 25, figsize= (12,6))
plt.show()

```

Here,we can visualize the distribution of length,depth and mass of our dataset.

# PAIR PLOT

```python
sns.pairplot(data = df,hue="species", height=3)
```

pairplot is showing the bivariate relation between each pair of features.
From the pairplot, we'll see that the Gento species is separataed from the other two across all feature combinations

# JOINT PLOT

```python
sns.jointplot(x="bill_length_mm", y="bill_depth_mm", data=df,height=5)
plt.show()
```

We can also use the seaborn library to make a similar plot.
A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure.

# VIOLIN PLOT

```python
sns.violinplot(x="species", y="body_mass_g", data=df, size=6)
plt.show()
```

Here,we can observe that both Adelie and Chinstrap species have almost equal range of body masses.
So,we can only differentiate the Gentoo species from the above violin plot.
We can also observe the outliers present among the Chinstrap species

# CONCLUSION

* The dataset is not balanced i.e. equal records are not present for all three species.
* We have four numerical columns while three categorical columns.
* A strong correlation is present between flipper length and body mass.
* The Gentoo species is the most easily distinguishable because of its large body masses.
* The Adelie and Chinstrap species are usually mixed and are sometimes hard to separate.
